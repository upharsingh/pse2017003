```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
from matplotlib import pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error,accuracy_score
from statistics import mean,median
```

```python
df=pd.read_csv("grngrass.csv")
#print(df)
arraylist=np.array(df)
meanof=np.mean(arraylist)
#meanofwave=np.mean(arraylist,axis=0)
meadianof=np.median(arraylist)
#meadianofwave=np.median(arraylist,axis=0)
minimum=np.min(arraylist)
maximum=np.max(arraylist)
varianceof=np.var(arraylist)
stdof=np.std(arraylist)
print("minimum is =",minimum)
print("maximum is =",maximum)
print("Mean is =",meanof)
#print("Mean is wavelength=",meanofwave)
print("Meadian is =",meadianof)
#print("Meadian is wavelength=",meadianofwave)
print("variance is =",varianceof)
print("standard deviation is =",stdof)
```

```python
features = ['wavelength','reflectance','standard deviation(bandwidth)']

# Separating out the features
x = df.loc[:, features].values
print(x)

# Separating out the target
#y = df.loc[:,['target']].values
#print(y)
```

```python
# Standardizing the features
x = StandardScaler().fit_transform(x)
print(x)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
ica = FastICA(n_components=2)
principalComponents2 = ica.fit_transform(x)
principalDf2 = pd.DataFrame(data = principalComponents2, columns = ['principal component 1', 'principal component 2'])
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PCA vs ICA', fontsize = 20)
ax.plot(principalDf['principal component 1'], principalDf['principal component 2'],color= 'Green')
ax.plot(principalDf2['principal component 1'], principalDf2['principal component 2'],color='Red')
plt.show()
```
